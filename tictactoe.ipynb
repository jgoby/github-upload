```python
#display current grid
def print_grid(grid):
    from IPython.display import clear_output
    clear_output()
    print(grid['1'])
    print(grid['2'])
    print(grid['3'])

#ask first player to choose symbol
def sym1_def():
    sym1=input('Player 1, would you like to be the X or O?: ').upper()
    while sym1 not in ['X','O']:
        sym1=input('Sorry, please choose an X or O: ').upper()
    print(f'Good, player 1, you are now an {sym1}.')
    return sym1

#set second player's symbol
def sym2_def(sym1):
    if sym1=='X':
        sym2='O'
    else:
        sym2='X'
    print(f'Player 2, this means you are an {sym2}.')
    return sym2

#ask for player choice
def pos(grid,turn,sym1,sym2,z1,z2,full_grid):
    pos=input(f'{turn}, please enter a grid spot, e.g. 2,1 for row 2 column 1: ')
    while pos not in full_grid:
        pos=input('Sorry, please enter in format of row 1, 2, or 3, comma, then column 1,2, or 3: ')
    while pos in z1 or pos in z2:
        pos=input(f'{turn}, please enter a grid spot that has not already been taken: ')
    column=int(pos[2])-1
    if turn=='Player 1':
        z1.append(pos)
        grid[pos[0]][column]=sym1
    elif turn=='Player 2':
        z2.append(pos)
        grid[pos[0]][column]=sym2   
    return grid,z1,z2

#when there is a winner, or cat's game (tie) with one blank position left, announce result
def check_for_win(turn,turn_on,full_grid,z1,z2,win1,win2,win3,win4,win5,win6,win7,win8):
    blanks=[]
    for i in full_grid:
        if i not in z1+z2:
            blanks.append(i)
    if all(x in z1 for x in win1) or \
    all(x in z1 for x in win2) or \
    all(x in z1 for x in win3) or \
    all(x in z1 for x in win4) or \
    all(x in z1 for x in win5) or \
    all(x in z1 for x in win6) or \
    all(x in z1 for x in win7) or \
    all(x in z1 for x in win8):
        print(f'{turn} wins!')
        turn_on=False
    elif all(x in z2 for x in win1) or \
    all(x in z2 for x in win2) or \
    all(x in z2 for x in win3) or \
    all(x in z2 for x in win4) or \
    all(x in z2 for x in win5) or \
    all(x in z2 for x in win6) or \
    all(x in z2 for x in win7) or \
    all(x in z2 for x in win8):
        print(f'{turn} wins!')
        turn_on=False   
    elif len(blanks)==1:
        if turn=='Player 1':
            aug2=z2+blanks
            if all(x in aug2 for x in win1) or \
            all(x in aug2 for x in win2) or \
            all(x in aug2 for x in win3) or \
            all(x in aug2 for x in win4) or \
            all(x in aug2 for x in win5) or \
            all(x in aug2 for x in win6) or \
            all(x in aug2 for x in win7) or \
            all(x in aug2 for x in win8):
                turn_on=True
            else:
                print("Cat's Game, i.e. tie!")
                turn_on=False
        if turn=='Player 2':
            aug1=z1+blanks
            if all(x in aug1 for x in win1) or \
            all(x in aug1 for x in win2) or \
            all(x in aug1 for x in win3) or \
            all(x in aug1 for x in win4) or \
            all(x in aug1 for x in win5) or \
            all(x in aug1 for x in win6) or \
            all(x in aug1 for x in win7) or \
            all(x in aug1 for x in win8):
                turn_on=True
            else:
                print("Cat's Game, i.e. tie!")
                turn_on=False
    return turn_on

#switch player
def switch_player(turn):
    if turn=='Player 1':
        turn='Player 2'
    elif turn=='Player 2':
        turn='Player 1'
    return turn
       
#ask if want to play again
def ask_to_play_again(game_on,turn_on):
    answer=input('Would you like to play again (Y or N)?: ').upper()
    while answer not in ['Y','N']:
        answer=input('Sorry, please choose a Y or N: ').upper()
    if answer=='Y':
        turn_on=True
    if answer=='N':
        game_on=False
    return game_on,turn_on

#actual function calls
def tictactoe():
    print('Welcome to tic tac toe!')
    game_on=True #resets game_on for another game
    turn_on=True #resets turn_on for another game
    win1=['1,1','2,1','3,1'] #defines all winning scenarios
    win2=['1,2','2,2','3,2']
    win3=['1,3','2,3','3,3']
    win4=['1,1','1,2','1,3']
    win5=['2,1','2,2','2,3']
    win6=['3,1','3,2','3,3']
    win7=['1,1','2,2','3,3']
    win8=['3,1','2,2','1,3']
    full_grid=win1+win2+win3 #defines full grid, and list of all possible grid choices
    while game_on==True: #continue this loop while players want to keep playing
        from IPython.display import clear_output
        clear_output() #clear display each loop cycle to keep clean
        grid={'1':[' ',' ',' '],'2':[' ',' ',' '],'3':[' ',' ',' ']} #resets grid rows
        z1=[] #resets list of player choices
        z2=[]
        turn='Player 1' #resets initial turn to Player 1
        sym1=sym1_def() #allows Player 1 to choose X or O symbol each time
        sym2=sym2_def(sym1)  
        
        while turn_on==True: #continue this loop while game is undecided
            pos(grid,turn,sym1,sym2,z1,z2,full_grid) #get player's grid choice
            print_grid(grid) # display updated grid
            turn_on=check_for_win(turn,turn_on,full_grid,z1,z2,win1,win2,win3,win4,win5,win6,win7,win8) #check for win or cat's game (tie)
            turn=switch_player(turn)
        game_on,turn_on=ask_to_play_again(game_on,turn_on) #turns off game_on if players don't want to continue, turns on turn_on if want to play again
    print('bye bye')
```
